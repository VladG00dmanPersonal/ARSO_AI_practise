```python
from scipy.optimize import minimize, LinearConstraint, NonlinearConstraint
import numpy as np
from tqdm import trange

def fun(x):
    # Если бы мы максимизировали просто произведение (попробуйте!), то
    # оптимизатор мог бы остановиться уже на первом шаге. Это связано
    # с тем, что произведение большого числа чисел меньших 1 очень близко
    # к нулю, поэтому оптимизатору может показаться, что какое значение ни
    # подставляй, целевая функция не меняется, то есть мы достигли локального
    # минимума! Чтобы этого не было, вместо произведение мы будем максимизировать
    # логарифм произведения (это то же самое, так как логарифм монотонно
    # возрастает). Логарифм произведения это сумма логарфимов.
    # Ну и наконец, так как мы максимизируем, мы ставим минус перед суммой
    # логарифмов.
    return -np.log(x).sum()

# Ограничение sum_i x_i = 1
constr1 = LinearConstraint(np.ones(10), 1, 1)

# Следующее ограничение можно сделать и с помощью LinearConstraint
def constr_fun(x):
    return (x * np.arange(1, 11)).sum()

constr2 = NonlinearConstraint(constr_fun, 5, 5)

# Строго говоря, у нас ограничение x_i >= 0, но для того, чтобы
# логарифм в fun работал корректно, мы отделяем допустимые значения
# от нуля, в надежде, что для оптимального значения все x_i будут
# больше 0.0001
bounds=[(0.0001, None) for _ in range(10)]

# Возьмем в качестве начального приближения вектор с равными значениями
# Нетрудно видеть, что для этого приближения не выполнено ограничение
# constr2. Часто в этом нет ничего страшного - в процессе работы оптимизатор
# может изменить x0 так, чтобы он удовлетворял всем ограничениям, но если
# ограничение будет довольно сложным, то у него может это не получиться,
# поэтому придется искать начальное приближение, которое удовлетворяет
# всем ограничениям
x0 = np.ones(10)/10

res = minimize(fun, x0=x0, bounds=bounds, constraints=[constr1, constr2])

res
```

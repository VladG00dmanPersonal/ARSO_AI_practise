```python
import numpy as np
from tqdm import trange

epoches = 10000
```

```python
np.random.randint(1, 7, 3)
```

```python
s = 0
e = 0 
for epoch in trange(epoches):
    a = np.sort(np.random.randint(1, 7, 3))

    if a[0] != a[1]:
        e += 1
        s += a[0]

s / e
```

```python
s = 0
for epoch in trange(epoches):
    cur = 0

    while cur < 1:
        cur += np.random.rand()
        s += 1

s / epoches
```
